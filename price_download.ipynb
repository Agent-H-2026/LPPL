```python
from schwabdev import client, enums, stream, tokens, translate
import os
import logging
import pandas as  pd
from dotenv import load_dotenv
load_dotenv()
```

```python
# set logging level
logging.basicConfig(level=logging.INFO)

#load environment variables and make client
client = client.Client(os.getenv('app_key'), os.getenv('app_secret'))
```

```python
ticker_list = ['/GC', 'GLD', '/SI', 'SLV','/BTC','/ETH','/CL','/NG']
FrequencyType = 'daily'
```

```python
if FrequencyType=='daily':
    out_dir = './data/daily'
elif FrequencyType=='weekly':
    out_dir = './data/weekly' 
elif FrequencyType=='monthly':
    out_dir = './data/monthly'
```

```python
for ticker in ticker_list:
    response = client.price_history(symbol= ticker,
    periodType= 'year',
    period= 20,
    frequencyType= FrequencyType,
    frequency= 1,
    needExtendedHoursData= False,
    needPreviousClose= True)
    df = pd.json_normalize(response.json()['candles'])
    df['datetime'] = pd.to_datetime(df['datetime'], unit='ms', utc=True)
    out_path = os.path.join(out_dir, f'{ticker.replace("/","_")}.parquet')
    df.to_parquet(out_path, engine='pyarrow', index=False)
    print(f'Saved {len(df)} rows to {out_path} ')
```
